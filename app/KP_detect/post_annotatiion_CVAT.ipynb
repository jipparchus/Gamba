## Remove unnecessary files created when copy the data in

```python
def remove_identifiter(path):
    """
    Remove 'ãƒ»Zone.Identifiter' files from the directory.
    Those files are created when files are copied from the host local machine.
    """
    files = os.listdir(path)
    for f in files:
        if f.endswith('.txt:Zone.Identifier'):
            os.remove(os.path.join(path, f))
            print('REMOVED:    ', f)
```

```python
remove_identifiter(app_sys.PATH_ASSET_KP_DETECT_ANNOTATED)
for i in ['Train', 'Test', 'Validation']:
    remove_identifiter(os.path.join(app_sys.PATH_ASSET_KP_DETECT_ANNOTATED, 'labels', i))
```

## Rename the images and corresponding annotation files (if not renamed before annotation)

This is just to ensure the file names to be unique

```python
def rename_files(target):
    new_names = []
    dict_new_old = {}
    path_target = os.path.join(app_sys.PATH_ASSET_KP_DETECT_ANNOTATED, f'{target}.txt')

    with open (path_target, 'r', encoding='utf-8') as f:
        datalist = f.readlines()
        for e, data in enumerate(datalist):
            data = data.split()
            img_name = os.path.split(data[0])[1]
            # BECAREFUL: extesion of the image file should be adjusted if pther format is used.
            img_name_new = f'{e:05d}.jpg'
            # Rename the image names in the txt file later
            new_names.append(f'{os.path.split(data[0])[0]}/{img_name_new}')
            dict_new_old[img_name_new] = img_name
            print('\n', img_name, '   =======>   ', img_name_new)
            # Find corresponding image names in the images directory and rename it
            img_full_path = os.path.join(app_sys.PATH_ASSET_KP_DETECT_ANNOTATED, 'images', target, img_name)
            img_full_path_new = os.path.join(app_sys.PATH_ASSET_KP_DETECT_ANNOTATED, 'images', target, img_name_new)
            os.rename(img_full_path, img_full_path_new)
            print(img_full_path, '\n=======>\n', img_full_path_new)
            # Find corresponding image names in the labels directory and rename it
            lbl_full_path = os.path.join(app_sys.PATH_ASSET_KP_DETECT_ANNOTATED, 'labels', target, img_name.replace('.jpg', '.txt'))
            lbl_full_path_new = os.path.join(app_sys.PATH_ASSET_KP_DETECT_ANNOTATED, 'labels', target, img_name_new.replace('.jpg', '.txt'))
            os.rename(lbl_full_path, lbl_full_path_new)
            print(lbl_full_path, '\n=======>\n', lbl_full_path_new)
    
    # Save the dictionary for future reference
    with open(os.path.join(app_sys.PATH_ASSET_KP_DETECT_ANNOTATED, f'{target}_new_old.json'), 'w') as f:
        json.dump(dict_new_old, f)

    os.rename(path_target, os.path.join(app_sys.PATH_ASSET_KP_DETECT_ANNOTATED, f'{target}_old.txt'))
    # Update the txt file
    with open (os.path.join(app_sys.PATH_ASSET_KP_DETECT_ANNOTATED, f'{target}.txt'), 'w', encoding='utf-8') as f:
        for i in new_names:
            f.write(i + '\n')
```

```python
# Rename images and corresponding text files
for tgt in ['Train', 'Test', 'Validation']:
    rename_files(tgt)
```
