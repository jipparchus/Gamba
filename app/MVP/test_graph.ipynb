# Solving PnP

```python
import cv2
```

```python
img_coords = np.array([list(coords) for coords in nx.get_node_attributes(GG, 'coords_2d').values()], dtype=np.float32)
object_coords = np.array([list(coords) for coords in nx.get_node_attributes(GG, 'coords_3d').values()], dtype=np.float32)
```

```python
fx, fy = 1000, 1000  # Estimated focal lengths (change based on your setup)
# cx, cy = 640, 360    # Assume the principal point is the image center
cx, cy = 0, 70    # Assume the principal point is the image center

camera_matrix = np.array([[fx, 0, cx],
                          [0, fy, cy],
                          [0,  0,  1]], dtype=np.float32)

# No lens distortion (assuming ideal pinhole camera)
dist_coeffs = np.zeros((4, 1), dtype=np.float32)
```

```python
success, rvec, tvec, inliers = cv2.solvePnPRansac(
    object_coords, 
    img_coords,
    camera_matrix,
    dist_coeffs,
    flags=cv2.SOLVEPNP_ITERATIVE,
    reprojectionError=8.0,
    confidence=0.99
)
```

```python
if success:
    print("Rotation Vector:\n", rvec)
    print("Translation Vector:\n", tvec)
    print("Inliers:\n", inliers)
else:
    print("PnP RANSAC failed.")
```

```python
T = np.eye(4)  # Initialize 4x4 identity matrix
R, _ = cv2.Rodrigues(rvec)  # Convert rotation vector to matrix
T[:3, :3] = R  # Set rotation
T[:3, 3] = tvec.flatten()  # Set translation
T
```

```python
npoints, ndim = object_coords.shape
M = np.ones((npoints, ndim+1))
M[:, :ndim] = object_coords
print(M)
```

```python
P = np.zeros((3, 4))
P[:, :3] = np.eye(3)
print(P)
```

```python
Q = camera_matrix @ P @ T
```

```python
points2d = Q @ M.T
points2d[:ndim-1, :].T
```

```python
for [x, y] in points2d[:ndim-1, :].T:
    plt.scatter(x, y)
```

```python
projected_points, _ = cv2.projectPoints(
    object_coords, rvec, tvec, camera_matrix, dist_coeffs
)
projected_points = projected_points.squeeze()  # Flatten (N, 1, 2)  (N, 2)
print(projected_points)
```

```python
h, w = 200, 300
image = np.ones((h, w, 3), dtype=np.uint8) * 255  # Create a blank white image
image2 = np.ones((h, w, 3), dtype=np.uint8) * 255  # Create a blank white image

# Draw original detected image points (in red)
for (x, y) in img_coords:
    cv2.circle(image, (int(x)+w//2, int(y)), 5, (0, 0, 255), -1)

# Draw projected 3D points (in green)
for (x, y) in projected_points:
    cv2.circle(image2, (int(x)+w//2, int(y)), 5, (0, 255, 0), -1)

# Show the image with the original and projected points
fig, (ax1, ax2) = plt.subplots(1,2)
ax1.imshow(image[:, :, ::-1])  # Convert BGR to RGB for Matplotlib
ax2.imshow(image2[:, :, ::-1])  # Convert BGR to RGB for Matplotlib
plt.show()
```
